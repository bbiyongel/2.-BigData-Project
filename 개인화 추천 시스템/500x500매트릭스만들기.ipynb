```python
import pandas as pd

result = pd.read_pickle("/content/drive/My Drive/Test 500x500 /resultrentlist.pkl")

tuser = pd.read_csv('/content/drive/My Drive/Test 500x500 /tuser4.csv',encoding = 'cp949',index_col = 0)
tuser
```

```python
# 매트릭스 만들 때 값으로 표현하기위해 모든 대출기록에 값 1을 추가
result['value'] = 1
```

```python
# 유저테이블을 대출건수 기준으로 내림차순으로 정렬(Top 10,50,100 읽을 수 있음)
tuser1 = tuser.sort_values(["대출건수"], ascending=[False])
# 500명의 대출기록을 빼는것보다 넉넉하게 빼야됨. 안그럼 매트릭스 만들 때 유저에서 빵꾸남.. 
tuser2 = tuser1.index[:600]
tuser2 = tuser2 + 1
```

```python
from tqdm import tqdm
# user 500위 이상 기록제거
result1 = result.copy()

tuser3 = tuser2.to_frame(index = 0 ,name = "회원번호")

# for liuser in tqdm(range(0,len(tuser2)), desc = "진행상황"):  
#         result1[liuser] 
#         result1 = result1[result1.회원번호 != tuser2[liuser]]       
# result1
```

```python
reuslt2 = pd.merge(tuser3,result1,on="회원번호")
```

```python
# 매트릭스 확인
matrix = pd.pivot_table(reuslt2,index='회원번호', columns='ISBN', values='value',fill_value = 0)
matrix.shape
```

```python
# 현재 뽑은 600명중에 가장 많이 빌린 500권을 추출함.
itemsum = matrix.sum(axis = 0, skipna = True) 

itemsort = itemsum.sort_values(ascending=False)

itemdf = itemsort.to_frame()

outitem = itemdf.index[:500]

outitemdf = outitem.to_frame(index = 0, name='ISBN')
```

```python
# item 대출건수기준 500위 이상 기록제거
reuslt3 = pd.merge(outitemdf,reuslt2,on="ISBN")
matrix = pd.pivot_table(reuslt3,index='회원번호', columns='ISBN', values='value',fill_value = 0)
matrix.shape
# for liitem in tqdm(range(0,len(outitem)), desc = "진행상황"):  
#         reuslt2 = reuslt2[reuslt2.ISBN != outitem[liitem]]       
# reuslt2.head()

# item 500개 완성
```

```python
# 유저도 초과된 인원을 300권을 대출한 횟수를 기준으로 300명을 추림.
usersum = matrix.sum(axis = 1, skipna = True) 

usersort = usersum.sort_values(ascending=False)

userdf = usersort.to_frame()

outuser = userdf.index[:500]

outuserdf = outuser.to_frame(index = 0, name='회원번호')
```

```python
real500x500 = pd.merge(outuserdf,reuslt3,on="회원번호")   
matrix = pd.pivot_table(real500x500,index='회원번호', columns='ISBN', values='value',fill_value = 0)
matrix.shape
# user 300명 완성
```
