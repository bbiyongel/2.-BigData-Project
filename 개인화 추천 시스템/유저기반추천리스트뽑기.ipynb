```
from tqdm import tqdm
prelist = {}
def getToplist(user,matrix,n):
  # 결과값 리스트선언
  booklist = []  
  sumboonmo = 0
  sumboonja = 0
  for title in tqdm(matrix.columns):
    for tuser in userdf.index :
       if user == tuser or float(userdf.loc[user][tuser]) < 0:
          continue   
       a = float(userdf.loc[user][tuser])      
       sumboonmo += a      
       sumboonja += a * matrix.loc[tuser,title]          
    booklist.append((title,'{:.5f}'.format(float(sumboonja/sumboonmo))))      
  booklist.sort(key=lambda r: r[1], reverse=True)        
  return booklist[:n]


 
```

```
 print(getToplist(126,matrix,10))
```
